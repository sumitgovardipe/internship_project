```python
from flask import Flask, request, jsonify
from flask_sqlalchemy import SQLAlchemy
from flask_cors import CORS
import threading
import requests
```

```python
app = Flask(__name__)
CORS(app)

app.config['SQLALCHEMY_DATABASE_URI'] = 'sqlite:///restaurant.db'
app.config['SQLALCHEMY_TRACK_MODIFICATIONS'] = False

db = SQLAlchemy(app)
```

```python
class MenuItem(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    name = db.Column(db.String(100))
    price = db.Column(db.Float)
    stock = db.Column(db.Integer)
```

```python
class Table(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    capacity = db.Column(db.Integer)
    is_available = db.Column(db.Boolean, default=True)
```

```python
class Order(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    item_id = db.Column(db.Integer)
    quantity = db.Column(db.Integer)
    total_price = db.Column(db.Float)
```

```python
class Reservation(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    table_id = db.Column(db.Integer)
    customer_name = db.Column(db.String(100))
```

```python
class Inventory(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    item_name = db.Column(db.String(100))
    quantity = db.Column(db.Integer)
    threshold = db.Column(db.Integer, default=5)
```

```python
with app.app_context():
    db.create_all()

print("✅ Database Created Successfully")
```

```python
with app.app_context():
    db.session.add_all([
        MenuItem(name="Pizza", price=250, stock=10),
        MenuItem(name="Burger", price=120, stock=20),
        Table(capacity=4),
        Table(capacity=2),
        Inventory(item_name="Cheese", quantity=50),
        Inventory(item_name="Buns", quantity=30)
    ])
    db.session.commit()

print("✅ Sample Data Inserted")
```

```python
@app.route('/menu', methods=['GET'])
def get_menu():
    items = MenuItem.query.all()
    return jsonify([
        {"id": i.id, "name": i.name, "price": i.price, "stock": i.stock}
        for i in items
    ])
```

```python
@app.route('/order', methods=['POST'])
def place_order():
    data = request.json
    item = MenuItem.query.get(data['item_id'])

    if not item or item.stock < data['quantity']:
        return jsonify({"error": "Item unavailable"}), 400

    total = item.price * data['quantity']
    item.stock -= data['quantity']

    order = Order(item_id=item.id, quantity=data['quantity'], total_price=total)
    db.session.add(order)
    db.session.commit()

    return jsonify({"message": "Order placed", "total": total})
```

```python
@app.route('/reserve', methods=['POST'])
def reserve_table():
    data = request.json
    table = Table.query.filter_by(is_available=True).first()

    if not table:
        return jsonify({"error": "No tables available"}), 400

    table.is_available = False

    reservation = Reservation(table_id=table.id, customer_name=data['name'])
    db.session.add(reservation)
    db.session.commit()

    return jsonify({"message": f"Table {table.id} reserved"})
```

```python
@app.route('/tables', methods=['GET'])
def view_tables():
    tables = Table.query.all()
    return jsonify([
        {"id": t.id, "capacity": t.capacity, "available": t.is_available}
        for t in tables
    ])
```

```python
@app.route('/sales', methods=['GET'])
def sales():
    orders = Order.query.all()
    total = sum(o.total_price for o in orders)

    return jsonify({
        "total_orders": len(orders),
        "total_sales": total
    })
```

```python
def run():
    app.run(port=5000)

thread = threading.Thread(target=run)
thread.daemon = True
thread.start()

print("✅ Server Started")
```
